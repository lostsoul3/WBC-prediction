```python
import numpy as np
from keras.optimizers import RMSprop
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, Lambda
from keras.layers import Dense
from keras.utils import np_utils
from keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import train_test_split
import cv2
import scipy
import os
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import json
import skimage.transform as st
```

```python
def get_json(foldername):
    path_to_json = foldername+'/'
    json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
    return json_files
```

```python
def get_json_data(foldername):
    labels = get_json(foldername)
    return labels
```

```python
def get_preprocessed_image(img):
    img = st.resize(img,(120, 160, 3))
    thr1 = cv2.dilate(img,None,iterations = 7)
    thr1 = cv2.erode(thr1,None,iterations = 10)
    img = np.asarray(thr1)
    return thr1
```

```python
def get_data(data_type, length):
    base_dir = 'isci archive dataset'
    name = base_dir+'/'+data_type+'/'
    X = []
    y = []
    count = 0
    for folder in os.listdir(name):
        foldername = name+str(folder)
        print foldername
        json_list = get_json_data(foldername)
        #print len(json_list)
        for list_value in json_list:
            json_name = foldername+'/'+list_value
            print json_name
            data = json.load(open(json_name))
            label = data['meta']['clinical']['benign_malignant']
            if label=='benign' and count<length:
                imagename = list_value.replace('.json','')
                imagename = foldername+'/'+imagename+'.jpg'
                print imagename
                img = cv2.imread(imagename)
                img = st.resize(img,(480, 640, 3))
                #img = get_preprocessed_image(img)
                img = np.asarray(img)
                X.append(img)
                y.append(str(label))
                count = count+1
            if label=='malignant':
                imagename = list_value.replace('.json','')
                imagename = foldername+'/'+imagename+'.jpg'
                print imagename
                img = cv2.imread(imagename)
                img = st.resize(img,(480, 640, 3))
                #img = get_preprocessed_image(img)
                img = np.asarray(img)
                X.append(img)
                y.append(str(label))
    X = np.asarray(X)
    y = np.asarray(y)
    return X,y
```

```python
X_train,y_train = get_data('train',173)

#print(y_train)
```

```python
X_train.shape
```

```python
unique=[]
for t in y_train:
    if t not in unique:
        unique.append(t)
print unique
```

```python
X_test, y_test = get_data('test',73)
```

```python
unique=[]

for t in y_test:
    if t not in unique:
        unique.append(t)
print unique
```

```python
encoder = LabelEncoder()
encoder.fit(y_train)
y_train = encoder.transform(y_train)
y_test = encoder.transform(y_test)
print len(y_train)
print len(y_test)
print X_train.shape
print X_test.shape
```

```python
def get_model():
    model = Sequential()
    model.add(Lambda(lambda x: x * 1./255., input_shape=(480, 640, 3), output_shape=(120, 160, 3)))
    model.add(Conv2D(4, (3, 3), input_shape=(480, 640, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(4, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(8, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
    model.add(Dense(8))
    model.add(Activation('relu'))
    model.add(Dropout(0.6))
    model.add(Dense(1))
    model.add(Activation('sigmoid'))
    rms = RMSprop(lr = 0.001,)
    model.compile(loss='binary_crossentropy',
                optimizer=rms,
                metrics=['accuracy'])

    return model
```

```python
model = get_model()
print(model.summary())
```

```python
epochs = 20
batch_size = 4
```

```python
history=model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=epochs,
    shuffle=True,
    batch_size=batch_size)

model.save_weights('balanced_model_2.h5')
```

```python
def plot_learning_curve(history):
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.savefig('./accuracy_curve.png')
    plt.clf()
    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.savefig('./loss_curve.png')

plot_learning_curve(history)
```

```python
from sklearn.metrics import accuracy_score

print('Predicting on test data')
y_pred = np.rint(model.predict(X_test))

print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```
